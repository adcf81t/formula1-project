```python
%sql
-- Databricks notebook source
CREATE DATABASE IF NOT EXISTS f1_raw;
```

```python
%sql
-- MAGIC %md
-- MAGIC #### Create tables for CSV files
desc database f1_raw
```

```python
%sql
DROP TABLE IF EXISTS f1_raw.circuits;
CREATE TABLE IF NOT EXISTS f1_raw.circuits(circuitId INT,
circuitRef STRING,
name STRING,
location STRING,
country STRING,
lat DOUBLE,
lng DOUBLE,
alt INT,
url STRING
)
USING csv
OPTIONS (path "/mnt/formula1dl4dbcourse/raw/circuits.csv", header true)
```

```python
%sql
SELECT * FROM f1_raw.circuits;
```
