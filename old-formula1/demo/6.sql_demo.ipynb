```sql
create or replace temp view v_drivers_standins_2018
as
select race_year, driver_name, team, total_points, wins, rank
from driver_standings
where race_year = 2018
```

```sql
create or replace temp view v_drivers_standins_2020
as
select race_year, driver_name, team, total_points, wins, rank
from driver_standings
where race_year = 2020
```

```sql
select * from v_drivers_standins_2020
```

```sql
select * from v_drivers_standins_2018
```

```sql
select * from v_drivers_standins_2018 d_2018
full join v_drivers_standins_2020 d_2020
on (d_2018.driver_name = d_2020.driver_name)
```
