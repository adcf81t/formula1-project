```python
# pyspark schema_of_json example

import pyspark.sql.functions as f

# Malformed JSON with unquoted field names
malformed_json = '{a: 1, b: "two"}'

# Infer schema, allowing unquoted field names
inferred_schema_options = f.schema_of_json(
    f.lit(malformed_json),
    {'allowUnquotedFieldNames': 'true'}
)

spark.range(1).select(inferred_schema_options.alias("schema_with_options")).show(truncate=False)
```

```python
%sql
SELECT schema_of_json('{"name": "John", "age": 30, "city": "New York"}') AS inferred_schema;
```

```python
%sql
SELECT from_json(
  '{"name": "John", "age": 30, "city": "New York"}',
  'name STRING, age BIGINT, city STRING'
) AS parsed_json;
```

```python
%sql
    
-- First, get the schema in a variable or directly
DECLARE OR REPLACE VARIABLE json_schema STRING;
SET VARIABLE json_schema = (SELECT schema_of_json('{"name": "John", "age": 30, "city": "New York"}'));

-- Then use it in from_json
SELECT from_json(
  json_column,
  'name STRING, age BIGINT, city STRING' -- Or use the variable here: ${json_schema}
) AS parsed_data
FROM your_table;
```
